```python
#import election_data.csv
#Please change the path of the election_data.csv before running

#read the election_data.csv file
#The current path on my machine is 'C:\\Users\\jhyan\\Documents\\all\\Data Science\\DS Boot Camp Northwestern\\Module 3 Challenge\\PyPoll\\Resources\\election_data.csv'
read = []

with open('C:\\Users\\jhyan\\Documents\\all\\Data Science\\DS Boot Camp Northwestern\\Module 3 Challenge\\PyPoll\\Resources\\election_data.csv', 'r') as myFile:
    lines=csv.reader(myFile)
    for line in lines:
        read.append(line)

#using threelists to store Date and Profit/Losses separately, head stores the header row
Ballot_ID = []
County = []
Candidate = []
head = read[0]

for i in range(1, len(read)):
    Ballot_ID.append(read[i][0])
    County.append(read[i][1])
    Candidate.append(read[i][2])

#The total number of votes cast
total_number = len(Ballot_ID)

#A complete list of candidates who received votes
candidates = Candidate[0]
candidates_list = []
candidates_list.append(Candidate[0])

for i in range(0, len(Candidate)):
    if Candidate[i] not in candidates_list:
        candidates_list.append(Candidate[i])

#The percentage of votes each candidate won and The total number of votes each candidate won
sum = 0
result_sum = []
result_average = []

for i in range(0, len(candidates_list)):
    for j in range(0, len(Candidate)):
        if Candidate[j] == candidates_list[i]:
            sum = sum + 1
    result_sum.append(sum)
    sum = 0
    
for i in range(0, len(result_sum)):
    result_average.append("{:.3%}".format(result_sum[i]/total_number))

#find the winner
max = result_sum[0]
max_name = candidates_list

for i in range(0, len(result_sum)):
        if result_sum[i] > max:
            max = result_sum[i]
            max_name = candidates_list[i]

#print all the results
print("Election Results")
print("----------------------------")
print("Total Votes:", total_number)
print("----------------------------")
print(candidates_list[0], ":", result_average[0], "(", result_sum[0], ")")
print(candidates_list[1], ":", result_average[1], "(", result_sum[1], ")")
print(candidates_list[2], ":", result_average[2], "(", result_sum[2], ")")
print("----------------------------")
print("Winner:", max_name)
print("----------------------------")

#write all the results in a txt file
fwrite = open('election_data_result.txt', 'w')
a = 'Election Results' + '\n'
b = '----------------------------' + '\n'
c = 'Total Votes:' + str(total_number) + '\n'
d = '----------------------------' + '\n'
e = str(candidates_list[0]) + ':' + str(result_average[0]) + '(' + str(result_sum[0]) + ')'+ '\n'
f = str(candidates_list[1]) + ':' + str(result_average[1]) + '(' + str(result_sum[1]) + ')'+ '\n'
g = str(candidates_list[2]) + ':' + str(result_average[2]) + '(' + str(result_sum[2]) + ')'+ '\n'
h = '----------------------------' + '\n'
i = 'Winner:' + max_name + '\n'
j = '----------------------------' + '\n'
fwrite.write(a)
fwrite.write(b)  
fwrite.write(c)
fwrite.write(d)
fwrite.write(e)
fwrite.write(f)
fwrite.write(g)
fwrite.write(h)
fwrite.write(i)
fwrite.write(j)
fwrite.close()
```
