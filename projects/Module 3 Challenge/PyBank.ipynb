```python
#import packages

import csv
```

```python
#import budget_data.csv file
#Please change the path of the budget_data.csv before running

#read the budget_data.csv file
#The current path on my machine is 'C:\\Users\\jhyan\\Documents\\all\\Data Science\\DS Boot Camp Northwestern\\Module 3 Challenge\\PyBank\\Resources\\budget_data.csv'
read = []

with open('C:\\Users\\jhyan\\Documents\\all\\Data Science\\DS Boot Camp Northwestern\\Module 3 Challenge\\PyBank\\Resources\\budget_data.csv', 'r') as myFile:
    lines=csv.reader(myFile)
    for line in lines:
        read.append(line)

#using two lists to store Date and Profit/Losses separately, head stores the header row
date = []
profit_losses = []
head = read[0]

for i in range(1, len(read)):
    date.append(read[i][0])
    profit_losses.append(int(read[i][1]))
    
#The total number of months
total_numberMon = len(date)

#The net total amount of "Profit/Losses" over the entire period
total_amountProLos = sum(profit_losses)

#the average of the changes in "Profit/Losses" over the entire period
change_amount = []
change_date = []
average_change = 0

for i in range(0, len(profit_losses)-1):
    change_amount.append(int(profit_losses[i+1] - profit_losses[i]))
    change_date.append(date[i+1])

average_change = round(sum(change_amount)/len(change_amount), 2)

#The greatest increase in profits (date and amount) over the entire period
increase = change_amount[0]
increase_date = 0

for i in range(0, len(change_amount)):
    if change_amount[i] > increase:
        increase = change_amount[i]
        increase_date  = change_date[i]

#The greatest decrease in profits (date and amount) over the entire period
decrease = change_amount[0]
decrease_date = 0

for i in range(0, len(change_amount)):
    if change_amount[i] < decrease:
        decrease = change_amount[i]
        decrease_date = change_date[i]

#print all the results
print("Financial Analysis")
print("----------------------------")
print("Total Months:", total_numberMon)
print("Total: $", total_amountProLos)
print("Average Change: $", average_change)
print("Greatest Increase in Profits:", increase_date, "($",increase, ")" )
print("Greatest Decrease in Profits:", decrease_date, "($", decrease,")")

#write all the results in a txt file
fwrite = open('budget_data_result.txt', 'w')
a = 'Financial Analysis' + '\n'
b = '----------------------------' + '\n'
c = 'Total Months:' + str(total_numberMon) + '\n'
d = 'Total: $' + str(total_amountProLos) + '\n'
e = 'Average Change: $' + str(average_change) + '\n'
f = 'Greatest Increase in Profits:' + str(increase_date) + '($' + str(increase) +  ')' + '\n'
g = 'Greatest Decrease in Profits:' + str(decrease_date) + '($' + str(decrease) +  ')' + '\n'
fwrite.write(a)
fwrite.write(b)  
fwrite.write(c)
fwrite.write(d)
fwrite.write(e)
fwrite.write(f)
fwrite.write(g)                                                           
fwrite.close()
```
